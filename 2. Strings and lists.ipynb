# **   F-**



 ( -   Python) f-,    format    .

```python
#   F-
x = 7
y = 10

print(f'  X = {x},   Y = {y}')

#   X = 7,   Y = 10
```

 f-   ,    :      ,    .
* : `{a:.xf}`,  a - , x -     .
*     : `{a:,}`,  a -  (   60000,  60,000).
*    : `{a:%}`,  a -  (    `{a:.0%}`).
*         (  e): `{a:e}`, `{a:2e}`  ..,  a - .
*   : `{a:d}`,   - , d -  .
*    , ..   ,        : `{a:d}`,  a - , d -  .
*  : `{a:<d}` -    , `{a:>d}` -    , `{a:^d}` -   ,  a - , d -    .

```python
#    f-
print(f'{2.12745:.2f}') # 2.13

#        f-
print(f'{60000:,}') # 60,000

#       f-
print(f'{0.5:%}') # 50.000000%
print(f'{0.5:.0%}') # 50%

#        f-
print(f'{1.2345678:e}') # 1.234568e+00
print(f'{1.2345678:.2e}') # 1.23e+00

#       f-
a = 12
print(f': {a:12}') # :           12

#    f-
print(f'{322:<20}') # 322                 
print(f'{322:>20}') #                  322
print(f'{322:^20}') #         322         
```
