# ** break, continue  else**

`break`    

`continue`     



`else` -  ,        (    false).        `break`  `else`  .



```python
#    break
i = 0

while True:
    i += 1
    if i == 1000:
        print(i) # 1000
        break

#    break, continue  else
a = ['', '', '', '']
x = 0

while x < len(a):
    if len(a[x]) >= 5: # ,       5 
        x += 1
        continue
    print("")
    break
else:
    print("") # 
```

 `range(start, stop, step)`       start  stop ( )   step (  1).

`range(5) == range(0, 5)`

 `enumerate()`      .

  `for i, j in enumerate(lst)`,  lst  ,  i   ,  j   .

```python
#   
for i in 'abcde':
    print(i, end=' ') # a b c d e 

#  ,      for,    
lst = [1, 2, 3, 4, 5]

for i in lst:
    i = 0

print(lst) # [1, 2, 3, 4, 5]

#       for (  )
lst = [1, 2, 3, 4, 5]

for i in range(len(lst)):
    lst[i] = 0

print(lst) # [0, 0, 0, 0, 0]

#    enumerate()
digs = [1, 3, -23, -32, 121, 15, 5]

for i, j in enumerate(digs):
    if 10 < abs(j) < 100:
        digs[i] = 0

print(digs) # [1, 3, 0, 0, 121, 0, 5]
```
