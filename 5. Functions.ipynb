```python
import time


#   
def get_nod(a, b):
    '''     a  b   
    :param a:   
    :param b:   
    :return: 
    '''
    while a != b:
        if a > b:
            a -= b
        else:
            b -= a
    return a


def test_nod(func):
    # ---  1 -----------
    a = 28
    b = 35
    res = func(a, b)
    if res == 7:
        print('#test1 - ok')
    else:
        print('#test1 - fail')


    # ---  2 -----------
    a = 100
    b = 1
    res = func(a, b)
    if res == 1:
        print('#test2 - ok')
    else:
        print('#test2 - fail')


    # ---  3 -----------
    a = 2
    b = 10000000
    st = time.time()
    res = func(a, b)
    et = time.time()
    dt = et - st
    if res == 2 and dt < 1:
        print('#test3 - ok')
    else:
        print('#test3 - fail')


test_nod(get_nod) # test1 - ok # test2 - ok # test3 - ok


#   
def get_nod(a, b):
    '''     a  b    
    :param a:   
    :param b:   
    :return: 
    '''
    if a > b:
        a, b = b, a
    while b != 0:
        a, b = b, a % b
    return a


def test_nod(func):
    # ---  1 -----------
    a = 28
    b = 35
    res = func(a, b)
    if res == 7:
        print('#test1 - ok')
    else:
        print('#test1 - fail')


    # ---  2 -----------
    a = 100
    b = 1
    res = func(a, b)
    if res == 1:
        print('#test2 - ok')
    else:
        print('#test2 - fail')


    # ---  3 -----------
    a = 2
    b = 10000000000000000000000000000000000000000000000000000
    st = time.time()
    res = func(a, b)
    et = time.time()
    dt = et - st
    if res == 2 and dt < 1:
        print('#test3 - ok')
    else:
        print('#test3 - fail')


test_nod(get_nod) # test1 - ok # test2 - ok # test3 - ok
```

```python
from functools import wraps

#   
def func_decorator(func):
    @wraps(func)
    def wrapper(s):
        return sum(func(s))
    
    return wrapper


@func_decorator
def get_list(s):
    '''     '''
    return list(map(int, s.split()))


print(get_list.__name__)
print(get_list.__doc__)
print(get_list('1 2 3 4 5'))
```
