```python
import time


#   
def get_nod(a, b):
    '''     a  b   
    :param a:   
    :param b:   
    :return: 
    '''
    while a != b:
        if a > b:
            a -= b
        else:
            b -= a
    return a


def test_nod(func):
    # ---  1 -----------
    a = 28
    b = 35
    res = func(a, b)
    if res == 7:
        print('#test1 - ok')
    else:
        print('#test1 - fail')


    # ---  2 -----------
    a = 100
    b = 1
    res = func(a, b)
    if res == 1:
        print('#test2 - ok')
    else:
        print('#test2 - fail')


    # ---  3 -----------
    a = 2
    b = 10000000
    st = time.time()
    res = func(a, b)
    et = time.time()
    dt = et - st
    if res == 2 and dt < 1:
        print('#test3 - ok')
    else:
        print('#test3 - fail')


test_nod(get_nod) # test1 - ok # test2 - ok # test3 - ok


#   
def get_nod(a, b):
    '''     a  b    
    :param a:   
    :param b:   
    :return: 
    '''
    if a > b:
        a, b = b, a
    while b != 0:
        a, b = b, a % b
    return a


def test_nod(func):
    # ---  1 -----------
    a = 28
    b = 35
    res = func(a, b)
    if res == 7:
        print('#test1 - ok')
    else:
        print('#test1 - fail')


    # ---  2 -----------
    a = 100
    b = 1
    res = func(a, b)
    if res == 1:
        print('#test2 - ok')
    else:
        print('#test2 - fail')


    # ---  3 -----------
    a = 2
    b = 10000000000000000000000000000000000000000000000000000
    st = time.time()
    res = func(a, b)
    et = time.time()
    dt = et - st
    if res == 2 and dt < 1:
        print('#test3 - ok')
    else:
        print('#test3 - fail')


test_nod(get_nod) # test1 - ok # test2 - ok # test3 - ok
```

```python
#       
def function(*args):
    print(args) # (1, 2, 3, 4, '5')


function(1, 2, 3, 4, '5')


#       
def function(**kwargs):
    print(kwargs) # {'a': 1, 'b': 2, 'number': True}


function(a=1, b=2, number=True)


#       
def function(n, *args, a=4, **kwargs):
    return n, *args, a, *kwargs.values()
 

print(function(1, 2, 3, b=5, c=6)) # (1, 2, 3, 4, 5, 6)
```
