```python
import time


#   
def get_nod(a, b):
    '''     a  b   
    :param a:   
    :param b:   
    :return: 
    '''
    while a != b:
        if a > b:
            a -= b
        else:
            b -= a
    return a


def test_nod(func):
    # ---  1 -----------
    a = 28
    b = 35
    res = func(a, b)
    if res == 7:
        print('#test1 - ok')
    else:
        print('#test1 - fail')


    # ---  2 -----------
    a = 100
    b = 1
    res = func(a, b)
    if res == 1:
        print('#test2 - ok')
    else:
        print('#test2 - fail')


    # ---  3 -----------
    a = 2
    b = 10000000
    st = time.time()
    res = func(a, b)
    et = time.time()
    dt = et - st
    if res == 2 and dt < 1:
        print('#test3 - ok')
    else:
        print('#test3 - fail')


test_nod(get_nod) # test1 - ok # test2 - ok # test3 - ok


#   
def get_nod(a, b):
    '''     a  b    
    :param a:   
    :param b:   
    :return: 
    '''
    if a > b:
        a, b = b, a
    while b != 0:
        a, b = b, a % b
    return a


def test_nod(func):
    # ---  1 -----------
    a = 28
    b = 35
    res = func(a, b)
    if res == 7:
        print('#test1 - ok')
    else:
        print('#test1 - fail')


    # ---  2 -----------
    a = 100
    b = 1
    res = func(a, b)
    if res == 1:
        print('#test2 - ok')
    else:
        print('#test2 - fail')


    # ---  3 -----------
    a = 2
    b = 10000000000000000000000000000000000000000000000000000
    st = time.time()
    res = func(a, b)
    et = time.time()
    dt = et - st
    if res == 2 and dt < 1:
        print('#test3 - ok')
    else:
        print('#test3 - fail')


test_nod(get_nod) # test1 - ok # test2 - ok # test3 - ok
```

```python
#   
def recursive(value):
    print(value, end=' ') #     2970,   RecursionError: maximum recursion depth exceeded while calling a Python object
    if value < 4: #    5,   0 1 2 3 4 
        recursive(value+1)


recursive(0)

#    
def recursive(value):
    print(value, end=' ')
    if value < 4: # 0 1 2 3 4
        recursive(value+1)
    print(value, end=' ') # 4 3 2 1 0 


recursive(0)


#   
def factorial(n):
    if n <= 0:
        return 1
    else:
        return n * factorial(n-1)
    

print(factorial(6)) # 720
```

# ** .   global  nonlocal**



** ** - ,        .

** ** -   .       .

    ,  ,    `global` .    ,         (). 

 `global`     ,      .

     - ,        ,  (  ) -     (  ).

  `nonlocal`        ( ,       ,         -    ).

```python
#  ,   ,  ,    
n = 12


def test():
    n = 10
    print(n) # 10


test()
print(n) # 12

#        
n = 12


def test():
    global n
    n = 20
    print(n) # 20


test()
print(n) # 20

#   nonlocal.    
x = 0


def outer():
    x = 1
    def inner():
        x = 2
        print('inner: ', x) # inner:  2

    inner()
    print('outer: ', x) # outer:  1


outer()
print('global; ', x) # inner:  2

#   nonlocal
x = 0


def outer():
    x = 1
    def inner():
        nonlocal x
        x = 2
        print('inner: ', x) # inner:  2

    inner()
    print('outer: ', x) # outer:  2

outer()
print('global; ', x) # inner:  2
```
